* is this difference a previous number that has been saved in the hashmap?
* if yes then you have your indices
* if know, record the new unseen number in your hashmap


# further details

* use hashmap 
* check if difference between target and an item in the array is already stored in the hashmap
* if the difference is not in the hashmap, then store the item that gave the difference with its index
* use enumerate()

```python
def twoSum(nums, target):
  # initiate hashmap
  SeenValues = {}
  for i, n in enumerate(nums):
    diff = target - n # take the diiference
    if diff in SeenValues:
      return [SeenValues[diff], i]
    SeenValues[n] = i # add the diff to the hashmap otherwise
  return 
```

# Try Example


```python
nums = [4, 6, 7, 712, 2,4, 4,0, 32, 55]
target = 38

twoSum(nums = nums, target = target)
```
