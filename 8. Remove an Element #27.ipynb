# Approach 1
* two pointers: One will be in the for loop. The other pointer L will be moving behind the R and collecting all trhe non-target values that R encounters on its way.
* Update L (to L + 1) to the next position to accept a potential next non-target number.
* return L

```python
# remove an element

class Solution:
  def removevalue(self, nums, val: int):
    l = 0
    for r in range(len(nums)):
      if nums[r] != val:
        nums[l] = nums[r]
        l = l + 1
    return l
```

```python
layers = [0, 1, 1, 1, 2, 2, 3, 4, 4, 7, 9, 10]
Solution().removevalue(nums = layers, val = 1)
```

# Approach 2

* use two pointers: one from left (L) and the other from the right(R)
* the idea is to move the non-targets to the left of the array and to move the target to the right of the array.
* the right pointer only moves inwards(to the left) if it lands on the target. (to keep the target on the right)
* it (R) moves until it lands on a non-target. Then it stays. It stays and waits until the left poiter finds a target to switch with. 
* in this case both the right and left pointer move inwards.
* assign the R's value to L and L's value to position R. Then move both L and R.


```python
class Solution:
  def removeval(self, nums, val):
    L, R = 0, len(nums) - 1
    c = 0
    while L <= R:
      if nums[R] == val:
        R = R - 1
        c = c + 1
      elif nums[L] != val:
        L = L + 1
      elif nums[L] == val:
        nums[R], nums[L] = nums[L], nums[R]
        L, R = L + 1, R - 1
        c = c + 1
    return len(nums) - c
```

# Approach 3
* while loop: along the array and remove the targets. 
* Else update the pointer to look at the next element in the array.
* return the length of the final array. 

```python
##  APPROACH 1 Repeat

class Solution:
  def RemoveNums(self, nums, target):
    i = 0
    while i < len(nums):
      if nums[i] == target:
        del nums[i]
      else:
        i += 1
    return len(nums)
```
