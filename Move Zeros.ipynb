# Move Zeros

# Approach 1
* Take steps from the end of the list
* start the pointer from the right
* if the element at the current position is a 0, then pop it out or delete it 

```python
def movezeros(nums):
  
  for j in range(len(nums)-1,-1,-1):
    if nums[j] == 0:
      del nums[j]
      nums.append(0)
      
  return nums
```

```python
movezeros([0,1,2,0,3,5,9,0,9,7])
```

```python
# edit move any number
# move zeros
class Solution:
  def MoveZeros(self, nums, target):
    for i in range(len(nums)-1, -1, -1):
      if nums[i] == target:
        del nums[i]
        nums.append(target)
    return nums
```

```python
movezeros([0,1,2,0,3,5,9,0,9,7])
```

# Approach 2
* Sliding window

```python
class Solution:
  def MoveZeros(self, nums):
    j = 0 
    for i in range(j, len(nums)):
      if nums[j] == 0:
        del nums[j]
        nums.append(0)
      else:
        j = j + 1
    return nums
```

```python
Solution().MoveZeros([1,2,3,4,0,0,9,53,0,0])
```
