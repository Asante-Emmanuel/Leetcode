# Remove Duplicates from a sorted array 
* Initiate two pointers at the second position.
* This is because the first position will always be unique
* Pointer 1 (L) keeps track of the next position to be filled by the next unique value. It moves up by 1 if only a new unique value is seen.
* Pointer 2 (R) moves along the array. it compares the current value to the previous value. 

## NB: Consider the case if the array is empty


```python
# Remove duplicates from an array of values
class Solution:
  def RemoveDuplicates(self, nums) -> int:
    l = 1 # initiates the pointers

    for r in range(1, len(nums)):
      if nums[r] != nums[r-1]:
        nums[l] = nums[r]
        l = l + 1 # update the left counter
    return l, nums


```

```python
Solution().RemoveDuplicates(nums = [0,0,1,1,1,2,2,3,3,4])
```
