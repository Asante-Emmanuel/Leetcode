# Approach 2 (Neetcode)

```python
class Solution:
  def RemoveNums(self, nums, target):
    L = 0
    for R in range(len(nums)):
      if nums[R] != target:
        nums[L] = nums[R]
        L = L + 1
    return L 
```

```python
Solution().RemoveNums(target = 4, nums = [1, 8, 9, 5, 6, 7, 2, 5, 5, 5, 3, 4, 4])
```

```python
class Solution:
  def RemoveNums(self, nums, target):
    L, R = 0, len(nums) - 1
    c = 0
    while L <= R:
      if nums[R] == target:
        R -= 1
        c += 1
      elif nums[L] != target:
        L += 1
      elif nums[L] == target:
        nums[L], nums[R] = nums[R], nums[L]
        L += 1
        R -= 1
        c += 1
    return  len(nums) - c
```

```python
Solution().RemoveNums(target = 5, nums = [1, 8, 9, 5, 6, 7, 2, 5, 5, 5, 3, 4, 4])
```

# Move Zeros

```python
def movezeros(nums):
  
  for j in range(len(nums)-1,-1,-1):
    if nums[j] == 0:
      del nums[j]
      nums.append(0)
      
  return nums

  
```

```python
movezeros([0,1,2,0,3,5,9,0,9,7])
```

```python
# move zeros
class Solution:
  def MoveZeros(self, nums, target):
    for i in range(len(nums)-1, -1, -1):
      if nums[i] == target:
        del nums[i]
        nums.append(target)
    return nums
```

```python
Solution().MoveZeros([1,2,3,5,9,0,0], target = 9)
```
