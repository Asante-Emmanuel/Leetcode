#3. Valid Palindrome
* Recognize to only focus on alphanumeric characters
* Compare the characters in lower case.
* Use "ord('a')" function to obtain the ASCII code for the alphanumeric characters
* The approach is to use two pointers. 
* Careful to consider the character on which the pointers are at during each step
* Use self.function to call internal function

```python
# define the function in 
class Solution:
  def isPalindrome(self, s:str) -> bool:
    l,r = 0, len(s) - 1
    
    while l < r:
      while l<r and not self.isAlphaNumeric(s[l]):
        l += 1
      while l<r and not self.isAlphaNumeric(s[r]):
        r -= 1
      if s[l].lower() != s[r].lower():
        return False
      l += 1; r -= 1
    return True
  
  # create a function for alphanumeric
  def isAlphaNumeric(self, c:str):
    return (ord('a') <= ord(c) <= ord('z') or 
            ord('A') <= ord(c) <= ord('Z') or 
            ord('0') <= ord(c) <= ord('9'))
```
