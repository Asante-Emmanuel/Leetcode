```python
# anagram
# compare lengths of strings
# compare the number each unique character appears in both strings (differ from neetcode)

class Solution:
  def isAnagram(self, s:str, t:str):
    if len(s) != len(t):
      return False
    
    # compare the counts
    #initiate the hashmaps
    countS, countT = {}, {}

    for i in range(len(s)):
      countS[s[i]] = 1 + countS.get(s[i], 0)
      countT[s[i]] = 1 + countT.get(s[i], 0)
    
    #return countS == countT


    for key in countS:
      if countS[key] != countT.get(key, 0):
        return False
    return True 
      

    
```

```python
# pick from A and drop from B

class Solution:
  def pickanddrop(self, s:str, t:str):
    if len(t) != len(s):
      return False
    # pick first character in s and check if it is in t
    n = len(t)
    i = 0
    while i < n:
      # find the first occurrence of s[i] in string t and drop it from string t
      if s[i] not in t:
        return False
      else:
        t = t[:t.index(s[i])] +  t[t.index(s[i])+1:]
        i = i + 1
    return True
        
```
