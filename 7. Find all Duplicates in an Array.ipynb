# Approach
* Notice that all entries of the array are between two values. i.e. a <= nums <= b
* Hence every entry - 1 is an index in the array.
* Iterate through the values and for each value, substract 1 from it and put the negated value at the index
* if the value at the index is negative, then add (index + 1 = the number) to the output array.

```python
class Solution():
  def FindDuplicates(self, nums):
    #initiate the output array
    output = []

    #iterate over the values in the array
    for i in range(len(nums) - 1):
      index = abs(nums[i]) - 1
      if nums[index] < 0:
        output.append(index + 1)
      nums[index] = -nums[index]
    
    return output
```

```python
Solution().FindDuplicates(nums = [4, 3, 2, 7, 8, 2, 3, 1])
```
