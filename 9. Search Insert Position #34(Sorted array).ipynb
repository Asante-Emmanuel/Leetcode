Given a sorted array of distinct integers and a target value, return the index if the target is found. If not, return the index where it would be if it were inserted in order.

You must write an algorithm with O(log n) runtime complexity.

* use binary search.
* initiate two pointers l,r  and midpoint
* compare the target to the midpoint and move the l or r (update the midpoint) accordingly to zoom in on the resulting subrange. 
* consider the cases in which the target falls out of the range of values in nums


```python
 # search insert position
class Solution:
  def SearchInsert(self, nums, target: int):
    l,r = 0, len(nums) - 1
    while l <= r:
      m = (r + l)//2
      if target == nums[m]:
        # update the upper limit (right index) and find the new midpoint
        return m
      elif target > nums[m]:
        # update the lower left index and find new midpoint
        l = m + 1
      else:
        r = m - 1
    return l
```

```python
Solution().SearchInsert(nums = [1,3,5,6], target = 2)
```
