#Approach #1

```python
# Approach 1

def isAnagram(s:str, t:str):
  if len(s) != len(t):
    return False

  # If it gets to this line of code, then the two are of equal length
  # create hashmaps for each string ( --> to compare their values later)

  countS, countT = {}, {} # initiate empty hashmaps

  # start the counter
  for i in range(len(s)):
    countS[s[i]] = 1 + countS.get(s[i], 0)
    countT[t[i]] = 1 + countT.get(t[i], 0)
  
  #compare the counts of elements of both strings
  for key in countS:
    if countS[key] != countT.get(key, 0):
      return False
  return True # At this point all the keys have been checked accross both hashmaps
    
```

####Try an Example

```python
s = 'denotes that there was an issue retrieving the key iskjl'
t = 'looking for. When you see a KeyError , the semantic mean'
isAnagram(s = s, t = t)
#isAnagram(s = s, t = sorted(s))
```

# Approach #2

```python
# use Counter function: creates hashmaps from strings
from collections import Counter

def isAnagram(s:str, t:str):
  return Counter(s) == Counter(t)
```

####Try an Example

```python
s = 'denotes that there was an issue retrieving the key iskjl'
t = 'looking for. When you see a KeyError , the semantic mean'
isAnagram(s = s, t = t)
#isAnagram(s = s, t = sorted(s))
```

#Approach #3

```python
# using sorting
# sort both strings and compare the sorted strings

def isAnagram(s:str, t:str):
  return sorted(s) == sorted(t)
```
