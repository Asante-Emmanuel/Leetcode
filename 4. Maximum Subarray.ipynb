```python
# maximum subarray

class Solution:
  def ismaxsub(self, nums):
    maxSub = 0
    cursum = nums[0]
    for n in nums:
      if cursum < 0:
        cursum = 0
      cursum  += n
      maxSub = max(maxSub, cursum)
    return maxSub
```

```python
solution = Solution()
solution.ismaxsub(nums = [-3, 4, 5 -2, 1, 9, 2, -4])
```
